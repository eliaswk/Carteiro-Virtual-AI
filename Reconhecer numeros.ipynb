```
# Abrindo imagem
from tensorflow.keras.preprocessing.image import load_img
img = load_img("1.png", grayscale=True)
```

```
# Reconhecendo numeros inscritos a m√£o
# Importando bibliotecas para redimensionar imagnes
from PIL import Image, ImageOps
import numpy  as np

img = img.resize((28,28)) # Redimensionando
img = ImageOps.invert(img) # invertendo imagem  
```

```
print(img.size)
```

```
# Mostrando imagem
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img, cmap="gray")
plt.show()
```

```
# Transfromando imagem em um array
from tensorflow.keras.preprocessing.image import img_to_array

img_array = img_to_array(img)
```

```
img_array.shape
```

```
# Aplicanto flattening
img_array = img_array.reshape((1,28 * 28))
```

```
# Normalizando imagem
img_array = img_array.astype("float32") / 255
```

```
# Carregando Modelo
from tensorflow.keras.models import load_model
model = load_model("Carteiro.h5")
```

```
predict = model.predict_classes(img_array)
```
